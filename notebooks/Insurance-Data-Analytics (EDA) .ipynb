```python
# Check outliers using IQR method

numerical_columns = ["Age","PremiumAmount","CoverageAmount","ClaimAmount"]

for column in numerical_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[
        (df[column] < lower_bound) |
        (df[column] > upper_bound)
    ]

    print(f"{column}: {len(outliers)} outliers")
```

Outlier Analysis

- Applied the IQR method to detect potential outliers in numerical columns.
- No outliers were identified in Age, PremiumAmount, CoverageAmount, or ClaimAmount.
- Therefore, no numerical records were removed during outlier treatment.

```python
# Claim Status by Policy Type

claim_policy = pd.crosstab(
    df["PolicyType"],
    df["ClaimStatus"]
)

claim_policy.plot(
    kind="bar",
    figsize=(10, 6)
)

plt.title("Claim Status by Policy Type")
plt.xlabel("Policy Type")
plt.ylabel("Number of Claims")
plt.xticks(rotation=0)
plt.legend(title="Claim Status")

plt.show()
```

# Claim Status by Policy Type Analysis

Key Insights:

- Rejected claims are the largest claim status category across all policy types.
- Travel insurance records the highest number of rejected, settled, and pending claims.
- Health insurance has the second-highest claim volume across the three claim status categories.
- Home insurance shows the lowest overall claim volume among the policy types.
- The high number of rejected Travel claims may be influenced by the larger number of Travel policies in the dataset.

```python
# Calculate Rejection Rate by Policy Type

rejection_rate = (
    df.groupby("PolicyType")["ClaimStatus"]
    .apply(lambda x: (x == "Rejected").mean() * 100)
    .sort_values(ascending=False)
)

print(rejection_rate)
```

```python
# Rejection Rate by Policy Type

plt.figure(figsize=(8, 5))

plt.bar(
    rejection_rate.index,
    rejection_rate.values,
    edgecolor="black"
)

plt.title("Rejection Rate by Policy Type")
plt.xlabel("Policy Type")
plt.ylabel("Rejection Rate (%)")

plt.show()
```

 Rejection Rate by Policy Type Analysis

Key Insights:

- Home insurance has the highest claim rejection rate among all policy types.
- Life insurance records the second-highest rejection rate.
- Travel insurance has the highest number of rejected claims, but it does not have the highest rejection rate.
- The high rejected claim count in Travel insurance is influenced by its larger policy volume.
- Home and Life insurance require closer investigation to understand the factors contributing to their higher rejection rates.

 Purpose:

- To compare claim rejection rates across policy types after accounting for differences in policy volume.
- To identify policy categories with a higher proportion of rejected claims.

```python
# Total Claim Amount by Policy Type

claim_amount_policy = (
    df.groupby("PolicyType")["ClaimAmount"]
    .sum()
    .sort_values(ascending=False)
)

print(claim_amount_policy)
```

```python
# Total Claim Amount by Policy Type

claim_amount_policy = (
    df.groupby("PolicyType")["ClaimAmount"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8, 6))

plt.pie(
    claim_amount_policy.values,
    labels=claim_amount_policy.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Claim Amount Share by Policy Type")

plt.show()
```

 Claim Amount Share by Policy Type Analysis

- To identify which policy category contributes the largest share of overall claim exposure.

```python
#Settled Claim Amount by Policy Type
settled_claim_amount = (
    df[df["ClaimStatus"] == "Settled"]
    .groupby("PolicyType")["ClaimAmount"]
    .sum()
    .sort_values(ascending=False)
)

print(settled_claim_amount)
```

```python
plt.figure(figsize=(9, 5))

plt.barh(
    settled_claim_amount.index,
    settled_claim_amount.values,
    edgecolor="black"
)

plt.title("Settled Claim Amount by Policy Type")
plt.xlabel("Total Settled Claim Amount")
plt.ylabel("Policy Type")

plt.gca().invert_yaxis()
plt.ticklabel_format(style="plain", axis="x")

plt.show()
```

 Settled Claim Amount by Policy Type Analysis

Key Insights:

- Travel insurance has the highest total settled claim amount among all policy types.
- Health insurance has the second-highest total settled claim amount.
- Home insurance records the lowest total settled claim amount.
- The higher settled claim amount for Travel insurance may be influenced by its larger number of policies and settled claims.
- Further analysis of the average settled claim amount is required to determine whether Travel claims are individually higher in value.

```python
# Average Settled Claim Amount by Policy Type

avg_settled_claim = (
    df[df["ClaimStatus"] == "Settled"]
    .groupby("PolicyType")["ClaimAmount"]
    .mean()
    .sort_values(ascending=False)
)

print(avg_settled_claim)
```

```python
# Average Settled Claim Amount by Policy Type

plt.figure(figsize=(9, 5))

plt.plot(
    avg_settled_claim.index,
    avg_settled_claim.values,
    marker="o",
    linestyle="None",
    markersize=10
)

plt.title("Average Settled Claim Amount by Policy Type")
plt.xlabel("Policy Type")
plt.ylabel("Average Settled Claim Amount")

plt.grid(axis="y", alpha=0.3)

plt.show()
```

Average Settled Claim Amount by Policy Type Analysis

Key Insights:

- Life insurance has the highest average settled claim amount among all policy types.
- Travel insurance has the highest total settled claim amount but does not have the highest average settled claim amount.
- This indicates that Travel's high total settled claim amount is primarily influenced by its larger settled claim volume.
- Home insurance records the lowest average settled claim amount.
- Life insurance may require closer financial risk monitoring due to its relatively higher average settled claim value.

```python
# Filter Settled Claims

settled_df = df[df["ClaimStatus"] == "Settled"]

# Premium Amount vs Settled Claim Amount

plt.figure(figsize=(9, 6))

plt.scatter(
    settled_df["PremiumAmount"],
    settled_df["ClaimAmount"],
    alpha=0.5
)

plt.title("Premium Amount vs Settled Claim Amount")
plt.xlabel("Premium Amount")
plt.ylabel("Settled Claim Amount")

plt.show()
```

 Premium Amount vs Settled Claim Amount Analysis

Key Insights:

- Claim amounts are widely distributed across different premium levels.
- No clear increasing or decreasing pattern is visually observed between Premium Amount and Settled Claim Amount.

```python
# Create Age Groups

def age_group(age):
    if age <= 30:
        return "Young"
    elif age <= 45:
        return "Adult"
    elif age <= 60:
        return "Middle Age"
    else:
        return "Senior"

df["AgeGroup"] = df["Age"].apply(age_group)

print(df["AgeGroup"].value_counts())
```

```python
# Claim Status by Age Group

age_claim_status = pd.crosstab(
    df["AgeGroup"],
    df["ClaimStatus"],
    normalize="index"
) * 100

print(age_claim_status)
```

```python
# Claim Status Percentage by Age Group

age_claim_status.plot(
    kind="bar",
    stacked=True,
    figsize=(10, 6)
)

plt.title("Claim Status Percentage by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Percentage of Claims (%)")
plt.xticks(rotation=0)
plt.legend(title="Claim Status")

plt.show()
```

 Claim Status Percentage by Age Group Analysis

Key Insights:

- Rejected claims represent the largest claim status proportion across all age groups.
- Young customers show the highest proportion of rejected claims compared with other age groups.
- Middle-aged customers show a relatively higher proportion of settled claims.
- Pending claim percentages remain relatively similar across the age groups.
- The analysis suggests that claim outcomes vary slightly across customer age segments.

```python
df["ClaimDate"] = pd.to_datetime(
    df["ClaimDate"],
    dayfirst=True,
    errors="coerce"
)

print(df["ClaimDate"].dtype)
```

```python
# Monthly Claim Amount Trend

monthly_claim = (
    df.dropna(subset=["ClaimDate"])
      .set_index("ClaimDate")
      .resample("ME")["ClaimAmount"]
      .sum()
)

print(monthly_claim)
```

```python
# Convert PolicyStartDate to Datetime

df["PolicyStartDate"] = pd.to_datetime(
    df["PolicyStartDate"],
    dayfirst=True,
    errors="coerce"
)

# Monthly Premium Trend

monthly_premium = (
    df.dropna(subset=["PolicyStartDate"])
      .set_index("PolicyStartDate")
      .resample("ME")["PremiumAmount"]
      .sum()
)

print(monthly_premium)
```

```python
# Plot Monthly Premium Trend

plt.figure(figsize=(11, 6))

plt.plot(
    monthly_premium.index,
    monthly_premium.values,
    marker="o"
)

plt.title("Monthly Premium Amount Trend")
plt.xlabel("Policy Start Month")
plt.ylabel("Total Premium Amount")

plt.ticklabel_format(style="plain", axis="y")
plt.grid(axis="y", alpha=0.3)
plt.xticks(rotation=45)

plt.show()
```

 Monthly Premium Amount Trend Analysis

Purpose:

- To analyze total premium amounts based on policy start month.
- To identify months with relatively high or low premium amounts.
- To examine fluctuations in premium amounts over time.

```python
# Create Coverage Amount Ranges

df["CoverageRange"] = pd.cut(
    df["CoverageAmount"],
    bins=[0, 30000, 50000, 70000, 90000, float("inf")],
    labels=[
        "Up to 30K",
        "30K-50K",
        "50K-70K",
        "70K-90K",
        "Above 90K"
    ]
)
```

```python
settled_df = df[df["ClaimStatus"] == "Settled"]
```

```python
plt.figure(figsize=(10, 6))

plt.boxplot(
    [
        settled_df[
            settled_df["CoverageRange"] == category
        ]["ClaimAmount"]
        for category in df["CoverageRange"].cat.categories
    ],
    tick_labels=df["CoverageRange"].cat.categories
)

plt.title("Settled Claim Amount Distribution by Coverage Range")
plt.xlabel("Coverage Amount Range")
plt.ylabel("Settled Claim Amount")

plt.xticks(rotation=20)

plt.show()
```

 Settled Claim Amount by Coverage Range Analysis

Key Insights:

- Settled claim amounts show a broadly similar distribution across all coverage ranges.
- The median settled claim amount remains relatively close across the coverage groups.
- Higher coverage ranges do not show a clear increase in typical settled claim amounts.
- The spread of settled claim amounts is also broadly similar across coverage categories.
- This suggests that Coverage Amount alone may not strongly explain differences in settled claim amounts in this dataset.

```python
# Claim Ratio by Policy Type

claim_ratio_policy = (
    df.groupby("PolicyType")
      .agg(
          TotalClaimAmount=("ClaimAmount", "sum"),
          TotalPremiumAmount=("PremiumAmount", "sum")
      )
)

claim_ratio_policy["ClaimRatio"] = (
    claim_ratio_policy["TotalClaimAmount"]
    / claim_ratio_policy["TotalPremiumAmount"]
)

claim_ratio_policy = claim_ratio_policy.sort_values(
    "ClaimRatio",
    ascending=False
)

print(claim_ratio_policy)
```

```python
# Claim Ratio by Policy Type

plt.figure(figsize=(9, 6))

plt.hlines(
    y=claim_ratio_policy.index,
    xmin=0,
    xmax=claim_ratio_policy["ClaimRatio"]
)

plt.plot(
    claim_ratio_policy["ClaimRatio"],
    claim_ratio_policy.index,
    "o",
    markersize=10
)

plt.title("Claim Ratio by Policy Type")
plt.xlabel("Claim Ratio")
plt.ylabel("Policy Type")

plt.show()
```

 Claim Ratio by Policy Type Analysis

Purpose:

- To compare the claim-to-premium ratio across different policy types.
- To identify policy categories with relatively higher claim amounts compared with premium amounts.#
- To evaluate differences in claim exposure across policy categories using a dataset-level ratio.
